```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import mysql.connector
```

```python
amazon_data=amazon_data.dropna()
amazon_data.shape
```

```python
#DATA CLEANING
amazon_data["discounted_price"]=amazon_data["discounted_price"].str.replace("₹","",regex=False).str.replace(",","",regex=False)
amazon_data["actual_price"]=amazon_data["actual_price"].str.replace("₹","",regex=False).str.replace(",","",regex=False)
amazon_data["discount_percentage"]=amazon_data["discount_percentage"].str.replace("%","",regex=False)
amazon_data["rating_count"]=amazon_data["rating_count"].str.replace(",","",regex=False)
amazon_data["rating"]=pd.to_numeric(amazon_data["rating"],errors="coerce")
```

```python
#Handling data types
amazon_data["actual_price"]=amazon_data["actual_price"].astype(float)
amazon_data["discounted_price"]=amazon_data["discounted_price"].astype(float)
amazon_data["discount_percentage"]=amazon_data["discount_percentage"].astype(int)
amazon_data["rating_count"]=amazon_data["rating_count"].astype(float)
amazon_data["rating_count"]= amazon_data["rating_count"].astype(int)
```

```python
amazon_data=amazon_data.dropna()
```

```python
category_types=amazon_data["category"]
print(list(category_types))
```

```python
categories_dataframe=pd.DataFrame(category_types)
print(categories_dataframe)
```

```python
def relevant_product(category_string):
    categories= category_string.split("|")
    relevant_category = categories[-1].strip().lower()
    return relevant_category
amazon_data["relevant_category"] =amazon_data["category"].apply(relevant_product)
print(amazon_data["relevant_category"])

```

```python

amazon_data["relevant_category"].unique()
```

```python
amazon_data=amazon_data.drop(columns=["category"])
amazon_data=amazon_data.rename(columns={"relevant_category":"category"})
amazon_data.columns
```

```python
#DATA TRANSFORMATION
amazon_data["discount_amount"]=amazon_data["actual_price"] -amazon_data["discounted_price"]
```

```python
amazon_data.to_csv("cleaned_amazon.csv")
```

```python
columns = []
for column, dtype in cleaned_amazon.dtypes.items():
    if dtype == "int64":
        columns.append(f"{column} INT")
    elif dtype == "float64":
        columns.append(f"{column} FLOAT")
    else:
        columns.append(f"{column} VARCHAR(255)")  

insert_query = f"""
    INSERT INTO products ({', '.join(cleaned_amazon.columns)})
    VALUES ({', '.join(['%s'] * len(cleaned_amazon.columns))});
"""
cursor=setup.cursor()
for index, row in cleaned_amazon.iterrows():
    try:
        cursor.execute(insert_query, tuple(row))
    except mysql.connector.Error as err:
        print(f"Error inserting {index}: {err}")

setup.commit()
cursor.close()
```

```python
#Top ten most reviewed products
mostreviewed_query="""SELECT product_id,product_name,rating_count FROM products
 ORDER BY rating_count DESC
LIMIT 10;
"""
most_reviewed10=pd.read_sql_query(mostreviewed_query,setup)
most_reviewed10.head(10)
```

```python
highest_category_rating_query="""SELECT category,AVG(rating) AS average_rating FROM products
 GROUP BY category ORDER BY average_rating DESC LIMIT 1;"""
highest_category_rating=pd.read_sql_query(highest_category_rating_query,setup)
highest_category_rating.head()
```

```python
discount_query="""SELECT product_id,product_name,discount_percentage FROM products WHERE discount_percentage > 50 ;"""
discount_50=pd.read_sql_query(discount_query,setup)
discount_50.head(10)
```

```python
most_reviewer_query="""SELECT user_id,user_name,COUNT(review_id) AS total_reviews FROM products 
GROUP BY user_id,user_name ORDER BY total_reviews DESC LIMIT 1;  """
highest=pd.read_sql_query(most_reviewer_query,setup)
highest.head()
```

```python
best_selling_query ="""SELECT category,SUM(rating_count) AS reviews
 FROM products GROUP BY category ORDER BY reviews DESC LIMIT 5 """
best_selling=pd.read_sql_query(best_selling_query,setup)
best_selling.head()
```

DATA ANALYSIS

```python
plt.figure(figsize=(10,6))
plt.title("RATINGS DISTRIBUTION")
sns.histplot(cleaned_amazon["rating"], bins=38, kde=True , color="blue")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```

```python
#correlation analysis between discount percentage and rating
correlation=cleaned_amazon["discount_percentage"].corr(cleaned_amazon["rating"])
print(f"The correlation between discount percentage and rating is:{correlation}")
```

```python
#Top selling product categories
reviews_per_category=cleaned_amazon.groupby("category")["rating_count"].sum().reset_index()
top_categories=reviews_per_category.sort_values(by="rating_count",ascending=False)
top_categories.head(10)
```

```python
correlation_discount_ratingcount=cleaned_amazon["discount_amount"].corr(cleaned_amazon["rating_count"])
print(correlation_discount_ratingcount)
```

Looking at the correlation results between discount amount and rating count which is (-0.04435431173801073),We find that discount amount has a weak negative correlation with rating count.That is as the discount amount increases the rating count slightly decreases.This relationship is not strong indicating that other factors might play a significant role in product rating count.
